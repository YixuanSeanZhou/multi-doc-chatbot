```python
from langchain.llms import LlamaCpp
from langchain.embeddings import LlamaCppEmbeddings
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter, RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
```

```python
llm = LlamaCpp(model_path="./models/llama-7b.ggmlv3.q4_0.gguf.bin", f16_kv=True)
```

```python
embeddings = LlamaCppEmbeddings(model_path="./models/llama-7b.ggmlv3.q4_0.gguf.bin", f16_kv=True)
```

```python
# load documents and create and store in vector database
from langchain.document_loaders import PyPDFLoader

pdf_loader = PyPDFLoader('./docs/RachelGreenCV.pdf')
documents = pdf_loader.load()
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=10)
documents = text_splitter.split_documents(documents)
len(documents)
```

```python
vectordb = Chroma.from_documents(
  documents,
  embedding=embeddings,
  persist_directory='./data'
)
vectordb.persist()
```

```python
import chromadb
client = chromadb.PersistentClient(path="./client_path")
```

```python
vectordb_loaded = Chroma(persist_directory='./data', embedding_function=embeddings)
```

```python
vectordb.embeddings
```

```python
vectordb_loaded.embedding
```

```python
from langchain.chains import RetrievalQA
chain = RetrievalQA.from_chain_type(llm=llm, 
                                    retriever=vectordb_loaded.as_retriever(search_kwargs={'k': 7}),
                                    return_source_documents=True
                                    )
```

```python
query = 'Who is the CV about?'
result = chain({'query': 'Who is the CV about?'})
result
```
